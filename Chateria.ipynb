

1.   Zadanie: Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka:

*   wypełnianie deklaracji PIT
 -  Wymaga znajomości zasad i znajomości przepisów podatkowych wymaganych do wypełnienia deklaracji PIT, nie wymaga kreatywności w podejściu ale wymaga wykorzystania uzyskanej wiedzy i zdolności logicznego myślenia, więc wymaga to inteligencji od człowieka.

*   streszczanie tekstu
 - Wymaga inteligencji, ponieważ trzeba zrozumieć treść, wyciągnąć kluczowe informacje i przeformułować je w bardziej zwięzły sposób. Jest to proces związany z rozumieniem języka i myśleniem abstrakcyjnym.

*   tłumaczenie tekstu
 - Tak, wymaga inteligencji, zwłaszcza jeśli chodzi o tłumaczenie kontekstowe i uwzględnianie niuansów językowych. Proste tłumaczenia mogą być automatyzowane, ale zaawansowane tłumaczenia nadal wymagają zrozumienia sensu i kulturowego kontekstu.

*   klasyfikacja tekstu do kategorii tematycznych
 - To zadanie wymaga inteligencji w sytuacji, gdy kategorie są nieoczywiste lub wymagają interpretacji kontekstu. Jednak w prostszych przypadkach można je zautomatyzować.

*   odpowiadanie na proste pytania zadawane w języku naturalnym (np. polskim)
 - To zadanie wymaga inteligencji ponieważ wymaga zrozumienia i znajomości języka naturalnego, jeżeli odpowiedź ma mieć jakikolwiek sens, jeżeli traktujemy odpowiadanie jako jakakolwiek reakcja to nie wymaga ona inteligencji

*   układanie rozkładu jazdy transportu miejskiego
 - Wymaga inteligencji, szczególnie jeśli trzeba uwzględniać złożone ograniczenia, takie jak godziny szczytu, potrzeby pasażerów i optymalizację kosztów.

*   programowanie (pisanie programów komputerowych)
 - Wymaga inteligencji, w tym logicznego myślenia, zdolności do rozwiązywania problemów i kreatywności.

*   "programowanie" kanałów telewizyjnych
 - Jeśli traktujemy programowanie jako zadanie polegające na ustawianiu kolejności programów na podstawie preferencji widzów i ograniczeń czasowych. Może wymagać pewnej analizy, i wymaga inteligencji od człowieka, a jeżeli chodzi o  szukanie dostępnych kanałów w telewizorze, za pośrednictwem ustawień w telewizorze to dalej wymaga to znajomości obsługi sprzętu który chcemy "zaprogramować"

*   testowanie oprogramowania
 - Wymaga inteligencji, zwłaszcza jeśli testy są projektowane w celu wykrycia trudnych do przewidzenia błędów.

*   komponowanie muzyki
 - Wymaga inteligencji, jest to zadanie bardzo kreatywne wymagające znajomości zasad które stosuje się w komponowaniu muzyki

*   rozwiązywanie układów równań
 - Wymaga inteligencji i zdolności logicznego myślenia oraz znajomości i zrozumienia zasad rozwiązywania układów równań, a jeżeli są bardziej skomplikowane to mogą wymagać też kreatywnego, nieodtwórczego podejścia do problemu

*   symboliczne obliczanie pochodnych funkcji
 - Wymaga inteligencji ponieważ wymaga znajomości obliczania pochodnych funkcji oraz w jaki sposób zapisać to symbolicznie.  

*   symboliczne całkowanie funkcji
 - symboliczne całkowanie wymaga inteligencji, ponieważ łączy w sobie umiejętność rozpoznawania wzorców, stosowania reguł, kreatywnego myślenia i elastycznego podejścia do problemów.

*   kierowanie samochodem
 - Wymaga inteligencji, ponieważ trzeba podejmować decyzje w dynamicznie zmieniających się warunkach, uwzględniając bezpieczeństwo i przepisy drogowe.

2.   Zadanie: Które z następujących problemów można uznać za mieszczące się w zakresie sztucznej
inteligencji:

*   streszczanie tekstu
 - znajduje się w zakresie sztucznej inteligencji, to zadanie wymaga przetwarzania języka naturalnego, przykładem modelu który jest w stanie przetwarzać język naturalny jest GPT

*   tłumaczenie tekstu
 - to zadanie również mieści się w zakresie AI, narzędzia takie jak Google Translate, Chat GPT są wstanie tłumaczyć tekst, uwzględniając kontekst, są też niestety często nadgorliwe przy próbach tłumaczenia i nie są tak elastyczne jak tłumaczenia wykonane przez człowieka ale dalej są w stanie tłumaczyć tekst

*   klasyfikacja tekstu do kategorii tematycznych
 - Tak, to zadanie z zakresu klasyfikacji danych, jest ono w zakresie uczenia maszynowego, na przykład przez wyszukiwanie słów kluczowych w tekście, i potrafią przypisać teksty do kategorii

*   odpowiadanie na proste pytania zadawane w języku naturalnym
 - Tak, AI potrafi interpretować pytania, potrafią przeanalizować kontekst oczywiście jeżeli taki zostanie podany lub jest na tyle oczywisty i wynikający z pytania. są też wstanie odpowiadać na takie pytania

*   rozwiązywanie układów równań
 - Tak, AI może to robić, szczególnie w przypadku systemów opartych na przetwarzaniu danych,to zadanie jest zautomatyzowane i zdeterminowane

*   układanie rozkładu jazdy
 - Tak, to zadanie dotyczące optymalizacji, AI może uwzględniać wiele zmiennych, takich jak czas, dostępne zasoby i potrzeby pasażerów.

*   rozwiązywanie układów równań liniowych
 - Tak, chociaż jest to często zautomatyzowane dzięki algorytmom, zadanie to wpisuje się w zakres AI, jeśli wymaga wyboru najlepszej metody lub adaptacji w złożonym środowisku.

*   symboliczne obliczanie pochodnych
 - Tak, Wymaga to interpretacji i zastosowania reguł matematycznych i leży to w zakresie AI

*   symboliczne całkowanie
 -Tak, Symboliczne całkowanie często wymaga wyszukiwania wzorców i stosowania różnych reguł, co leży w obszarze możliwości AI

*   kierowanie samochodem
 - Tak, chociaż jest to złożona kwestia, zarówno z perspektywy moralnej jak i technicznej, ale leży ono w zakresie AI, wymaga połączenia technik widzenia komputerowego, podejmowania decyzji, uczenia maszynowego oraz planowania w czasie rzeczywistym.

3. Zadanie: Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

* rozmowa towarzyska
 - może być skutecznie imitowana, tak długo jak rozmowa jest dość prosta, nie uwzględnia niszowych tematów lub nie wykracza poza zakres wiedzy i znajomości bazy danych do której ma dostęp sztuczne systemy, jednak jeżeli chcemy poprowadzić głęboką filozoficzną dyskusję oprogramowanie może nie spełniać tej roli tak skutecznie

* dyskusja polityczna
 - AI może służyj jako pewnego rodzaju tester, narzędzie które pozwala weryfikować fakty, oraz weryfikować wypowiedzi polityków i ich zgodność z prawdą i stanem faktycznym, więc jeżeli spojżymy na dyskusję polityczną jako weryfikację faktów i działań które podejmują politycy, to tak, jeżeli ma imitować dyskusję na tematy polityczne to pewnie wyjdzie ona bardziej sztucznie niż dyskusja z człowiekiem, ze względu na niespójność ludzkich poglądów, AI może być bardziej spójne w swoich wypowiedziach, lub być tak samo niespójne jak wypowiedzi człowieka, więc myślę że pod tym względem mogą być skuteczne ale nie ma takiej gwarancji

* dyskusja naukowa
 - AI nie ma zdolności do krytycznego myślenia, nie jest w stanie wyciągnąć własnych, orginalnych wniosków, może próbować prowadzić dyskusję na temat który zostanie zaproponowany jeżeli otrzyma dane, ale nie jest kompetentna jeśli chodzi o takie działanie

* odpowiadanie na pytania klientów w telefonicznej infolinii
 - asystenci głosowi i system IVR potrafią udzielić odpowiedzi głosowej i mogą skutecznie odpowiadać na pytania klientów, szczególnie jeśli dotyczą one powtarzalnych problemów, jeżeli problem jest nietypowy lub złożony lub jeżeli klient mówi z akcentem lub mówi szybko i niewyraźnie pojawiają się problemy z otrzymaniem odpowiedzi na pytanie.

* odpowiadanie na pytania klientów w internetowej infolinii
 - Chatboty internetowe są bardzo skuteczne w udzielaniu odpowiedzi na pytania klientów, szczególnie gdy pytania są jasno sformułowane. Mogą korzystać z baz wiedzy i algorytmów przetwarzania języka naturalnego, aby dostarczać pomoc w czasie rzeczywistym, jednak jeżeli problem jest bardziej złożony, mogą one nie spełniać swojej roli w wystarczającym zakresie.

  1. Przeprowadź rozmowę z chatbotem. Spróbuj zdefiniować różnice pomiędzy
botem udającym człowieka (przygotowywanym na test Turinga) a botem
„asystentem, służącym”.

 - różnica między botem udającym człowieka a asystentem służącym to fakt że boty udające człowieka starają się imitować zachowanie i spontaniczne reakcje człowieka, stara się imitować emocje i wydawać się przy tym jak najbardziej naturalnym, uwzględniając niuanse językowe, może stosować żarty językowe aby dodać sobie więcej wiarygodności aby przejść test Turinga, natomiast asystent lub bot służący ma na celu być narzędziem ułatwiającym funkcjonowanie, uproszczającym czynności, nie próbuje on udawać człowieka ponieważ nie jest to potrzebne aby był użyteczny/funkcjonalny, dodatkowo, mogłoby to utrudnić działanie i komunikację asystent-użytkownik.

 3. Raport:

  testowane były boty ze strony https://www.cleverbot.com oraz https://chatbot.pl

- a) Opowiadanie żartów

chatbot.pl: Posiada zestaw odtwórczych żartów, które są raczej proste i zaczerpnięte z zasobów zewnętrznych. Może opowiadać żarty, jeśli użytkownik zainicjuje interakcję w tym zakresie.

Cleverbot: W języku polskim zna jeden żart, natomiast w języku angielskim twierdzi, że nie zna żadnych żartów, po czym „domaga się” ich od użytkownika. Jego podejście do humoru jest ograniczone i chaotyczne.

- b) Przytaczanie cytatów z twoich wypowiedzi, lub znanych osób

chatbot.pl i Cleverbot: Nie potrafią przytaczać wcześniejszych wypowiedzi ani cytatów znanych osób. Brakuje im pamięci kontekstowej i zdolności odniesienia się do wcześniejszych etapów rozmowy.

- c) Nawiązywanie wypowiedzi do słów kluczowych

chatbot.pl: Działa w dużej mierze na podstawie słów kluczowych, reagując na nie według przypisanej hierarchii. Nie przetwarza złożonych wiadomości, jeśli brakuje w nich odpowiednich fraz.

Cleverbot: Nie bazuje na słowach kluczowych w tak prosty sposób jak chatbot.pl, ale jego odpowiedzi są chaotyczne i często oderwane od tematu rozmowy.

- d) Zadawanie dużej liczby pytań

chatbot.pl: W przypadku braku zrozumienia pytania lub zdania, generuje listę pytań, na które może odpowiedzieć, zamiast angażować się w rozmowę.

Cleverbot: Może zadawać pytania, ale często są one przypadkowe lub nie na temat, co sprawia wrażenie braku konsekwencji w rozmowie.

- e) Powracanie do początku wypowiedzi, sekwencyjne powtarzanie

chatbot.pl: Nie przejawia takich zachowań, ponieważ działa na prostych regułach i słowach kluczowych.

Cleverbot: Nie wykazuje powtarzania całych wypowiedzi, ale często wraca do niespójnych wątków lub zmienia kierunek rozmowy bez logicznego uzasadnienia.

- f) Zadawanie pytań powstających z twoich wypowiedzi

chatbot.pl: Nie wykazuje zdolności do generowania pytań na podstawie rozmowy – odpowiada wyłącznie na pytania zgodne z zaprogramowanymi regułami.

Cleverbot: Czasami generuje pytania na podstawie wprowadzonych danych, ale są one często przypadkowe i nie wynikają z kontekstu rozmowy.

- g) Odpowiadanie wymijająco, ogólnikowo

chatbot.pl: Odpowiada wymijająco w przypadku braku rozpoznania słów kluczowych, kierując rozmowę na tematy, na które może odpowiedzieć.

Cleverbot: Jego odpowiedzi są często ogólnikowe lub całkowicie oderwane od pytania, co może wynikać z braku zrozumienia treści rozmowy.

- h) Częsta zmiana tematu rozmowy

chatbot.pl: Raczej nie zmienia tematu – jego odpowiedzi są zaprogramowane i ograniczone do obsługi zapytań zgodnych z regułami.

Cleverbot: Często zmienia temat, co wynika z jego chaotycznego sposobu generowania odpowiedzi i braku pamięci kontekstowej.

- i) Problemy z utrzymaniem wątków

chatbot.pl: Nie próbuje utrzymać wątków – reaguje na każde zapytanie niezależnie, co powoduje, że rozmowa traci ciągłość.

Cleverbot: Ma duże problemy z utrzymaniem wątków – jego odpowiedzi są niespójne i często nie odnoszą się do wcześniejszych części rozmowy.

- Podsumowanie:

chatbot.pl: Oparty na regułach i słowach kluczowych, nie angażuje się w dynamiczną rozmowę, jest przewidywalny, ale ograniczony, przypomina wersję demonstracyjną.

Cleverbot: Stara się imitować ludzką rozmowę, ale wykazuje chaotyczne zachowanie, problemy z logiką i brak pamięci kontekstowej. Oba boty mają istotne ograniczenia, które sprawiają, że rozmowa z nimi szybko ujawnia ich sztuczną naturę.
